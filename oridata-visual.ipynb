```python
import jax
import jax.numpy as jnp

data = jnp.load('data/mnist.npz')
x_train, y_train, x_test, y_test = data.values()
```

pca visual

```python
import pcax
import matplotlib.pyplot as plt

# 1. 建立PCA对象，降到2维
x_train_embedded = pcax.fit(x_train.reshape(x_train.shape[0],-1),n_components=2)
x_train_embedded = pcax.transform(x_train_embedded, x_train.reshape(x_train.shape[0],-1))

# 可视化 PCA 结果
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_train_embedded[:, 0], x_train_embedded[:, 1], c=y_train, cmap='viridis', alpha=0.6)
plt.colorbar(scatter)  
# 添加颜色条，显示标签的映射
plt.title("PCA Visualization of Global Dataset")
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.show()
```

tsne visual

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# 使用t-SNE进行降维
x_train_embedded = TSNE(n_components=2, random_state=42).fit_transform(x_train.reshape(x_train.shape[0],-1))

# 可视化 t-SNE 结果
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_train_embedded[:, 0], x_train_embedded[:, 1], c=y_train, cmap='viridis', alpha=0.6)
plt.colorbar(scatter)  
# 添加颜色条，显示标签的映射
plt.title("t-SNE Visualization of Global Dataset")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.show()
```
